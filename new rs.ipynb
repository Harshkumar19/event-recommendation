```python
import pandas as pd
import numpy as np
```

```python
df_groups = pd.read_csv('meta-groups.csv')
```

```python
df_events= pd.read_csv('meta-events2.csv')
```

```python
df_mted = pd.read_csv('member-to-group-edges.csv')
```

```python
df_new= pd.merge(df_events, df_groups, on=['group_id'])
```

```python
agg_functions = {'group_id': 'first', 'weight': 'sum'}
```

```python
df_mted_new = df_mted.groupby(df_mted['group_id']).aggregate(agg_functions)
```

```python
df_mted_new['weight'].unique()
```

```python
# df_mted_new.drop(columns=['group_id'])
```

```python
# df_mted_new = df_mted_new.reset_index()
```

```python
df_final = pd.merge(df_new, df_mted_new['weight'], on=['group_id'])
```

```python
df_final = df_final[['event_id','group_id','name','time','group_name','num_members','category_name','weight']]
```

```python
df_final.info()
```

```python
df_final['time']=pd.to_datetime(df_final['time'])
```

```python
df_final['year'] = df_final['time'].dt.year
df_final['month'] = df_final['time'].dt.month
df_final['dow'] = df_final['time'].dt.day_of_week
```

```python
df_final = df_final.drop(columns=['time','group_id'])
```

```python
df_final
```

```python
df_final['year'].nunique()
```

```python
df_final['num_members'] = df_final['num_members'].apply(np.str)
df_final['weight'] = df_final['weight'].apply(np.str)
df_final['year'] = df_final['year'].apply(np.str)
df_final['month'] = df_final['month'].apply(np.str)
df_final['dow'] = df_final['dow'].apply(np.str)
```

```python
df_final['tags'] = df_final['group_name'] + df_final['num_members'] + df_final['category_name'] + df_final['weight'] + df_final['year'] + df_final['month'] + df_final['dow']
```

```python
df_final['tags'][0]
```

```python
df_final = df_final[['event_id','name','tags']]
```

```python
df_final
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(df_final['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
df_final[df_final['name'] == 'Steak Dinner on the Patio'].index[0]
```

```python
def recommend(event):
    index = df_final[df_final['name'] == event].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df_final.iloc[i[0]]['name'])
#         print('and tags are')
#         print(df_final.iloc[i[0]]['tags'],'\n')
        
```

```python
recommend('Steak Dinner on the Patio')
```
